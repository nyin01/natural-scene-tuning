```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('../images/original/test0.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
'''
threshold1: 
    Pixels with gradient values higher than this threshold are considered edges.
threshold2: 
    Pixels with gradient values between threshold1 and threshold2 are included as edges if they are connected to strong edges 
    (i.e., pixels with gradient values greater than threshold2).
apertureSize: 
    Aperture size for the Sobel operator. It determines the size of the kernel used for gradient computation. By default, it is 3.
L2gradient: 
    If True, the function uses the L2 gradient magnitude for the edge detection.
    If False, it uses the L1 gradient magnitude. By default, it is False.
'''
edges = cv.Canny(img, threshold1=50, threshold2=20)
```

```python
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
